```python
plt.hist(np.random.exponential(1,100000), density=True, bins=50, histtype="step", color="blue",
label="exp")
plt.hist(np.random.normal(0,1,100000), density=True, bins=50, histtype="step", color="red",
label="normal")
plt.hist(np.random.uniform(0,1,100000), density=True, bins=50, histtype="step", color="green",
label="uniform")
plt.axis([-8,8,0,1.2])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.figure(1, figsize=(24, 8))
subplot3d = plt.subplot(111, projection='3d')
subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
stats_df = df.drop(['Name', 'Type 1', 'Type 2', 'Total', 'Stage', 'Legendary'], axis=1)
stats_df.head()
```

```python
pearson_corr = stats_df.corr()
pearson_corr
```

```python
spearman_corr = stats_df.corr('spearman')
spearman_corr
```

```python
figure, axes = plt.subplots(2,1, figsize=(9,16))

pearson_subplot = sns.heatmap(pearson_corr, annot=True, ax=axes[0])
pearson_subplot.set_title('Pearson Correlation', fontsize=18)

spearman_subplot = sns.heatmap(spearman_corr, annot=True, ax=axes[1])
spearman_subplot.set_title('Spearman Correlation', fontsize=18)

plt.show()
```
